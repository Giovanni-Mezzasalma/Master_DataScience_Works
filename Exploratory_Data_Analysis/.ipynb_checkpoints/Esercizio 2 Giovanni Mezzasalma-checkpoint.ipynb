````python
# Import Libraries
import numpy as np
import pandas as pd
from urllib.request import urlopen  
import os.path as osp
import os
import logging
import zipfile
from glob import glob
from random import randint
import matplotlib.pyplot as plt

logging.getLogger().setLevel('INFO')
````

## Helpers

````python
def download_file(url_str, path):
    url = urlopen(url_str)
    output = open(path, 'wb')       
    output.write(url.read())
    output.close()  
    
def extract_file(archive_path, target_dir):
    zip_file = zipfile.ZipFile(archive_path, 'r')
    zip_file.extractall(target_dir)
    zip_file.close()
````

## Download the dataset

````python
BASE_URL = 'http://tennis-data.co.uk'
DATA_DIR = "tennis_data"
ATP_DIR = './{}/ATP'.format(DATA_DIR)
WTA_DIR = './{}/WTA'.format(DATA_DIR)

ATP_URLS = [BASE_URL + "/%i/%i.zip" % (i,i) for i in range(2000,2019)]
WTA_URLS = [BASE_URL + "/%iw/%i.zip" % (i,i) for i in range(2007,2019)]

os.makedirs(osp.join(ATP_DIR, 'archives'), exist_ok=True)
os.makedirs(osp.join(WTA_DIR, 'archives'), exist_ok=True)

for files, directory in ((ATP_URLS, ATP_DIR), (WTA_URLS, WTA_DIR)):
    for dl_path in files:
        logging.info("downloading & extracting file %s", dl_path)
        archive_path = osp.join(directory, 'archives', osp.basename(dl_path))
        download_file(dl_path, archive_path)
        extract_file(archive_path, directory)
    
ATP_FILES = sorted(glob("%s/*.xls*" % ATP_DIR))
WTA_FILES = sorted(glob("%s/*.xls*" % WTA_DIR))

df_atp = pd.concat([pd.read_excel(f) for f in ATP_FILES], ignore_index=True)
df_wta = pd.concat([pd.read_excel(f) for f in WTA_FILES], ignore_index=True)

logging.info("%i matches ATP in df_atp", df_atp.shape[0])
logging.info("%i matches WTA in df_wta", df_wta.shape[0])
````

````python
#extraction of the years in which a player has played the first 
#(and the last) game (respectively) in the considered timeframe

player_w=df_atp['Winner'].unique()
player_l=df_atp['Loser'].unique()

year_loser_min=[]
year_winner_min=[]

for player in player_w:
    min_w=df_atp[df_atp['Winner']==player]['Year'].unique().min()
    year_winner_min.append(min_w)

df_player_w=pd.DataFrame([player_w, year_winner_min]).T

for player in player_l:
    min_l=df_atp[df_atp['Loser']==player]['Year'].unique().min()
    year_loser_min.append(min_l)

df_player_l=pd.DataFrame([player_l, year_loser_min]).T
````

````python
#concatenation of the obtained data e renaming of columns

players=pd.concat([df_player_w,df_player_l], ignore_index=True)
players = players.rename(columns={
    0: 'Name',
    1: 'First_year',
})
````

````python
#chaging of the column 'First_year' dtype for the approximation of the players actual age
#and dropping the duplicated values keeping the significative ones

players= players.astype({"First_year": int})

starting_age=19

players['Age_actual']=2022-players['First_year']+starting_age
index=players['Age_actual'].argsort()
players=players.iloc[index]
players=players.drop_duplicates(subset='Name', keep="last")
````

````python
#Calculation of the players (winner and loser) age when the game was played

df_atp['Age_winner']=b
df_atp= df_atp.astype({"Age_winner": int})
df_atp['Age_winner']=df_atp['Age_winner']-(2022-df_atp['Year'])

df_atp['Age_loser']=d
df_atp= df_atp.astype({"Age_loser": int})
df_atp['Age_loser']=df_atp['Age_loser']-(2022-df_atp['Year'])
````

4. For each match, what is the percentage of victories of the winner in the past ?

````python
#looking for memory usage and for missing values

df_atp.info()
````

````python
perc_win={player:[] for player in players['Name']}
perc_lose={player:[] for player in players['Name']}

match={player:0 for player in players['Name']}
win={player:0 for player in players['Name']}
lose={player:0 for player in players['Name']}

matches={player:[] for player in players['Name']}

for winner,loser in zip(df_atp['Winner'],df_atp['Loser']):
    match[winner]+=1
    match[loser]+=1
    win[winner]+=1
    lose[loser]+=1
    matches[winner].append(match[winner])
    matches[loser].append(match[loser])
    perc_win[winner].append((win[winner])/match[winner])
    perc_lose[loser].append(lose[loser]/match[loser])
````

````python
%%time
ind_win={player:[] for player in players['Name']}
ind_lose={player:[] for player in players['Name']}

for player in players['Name']:
    ind_win[player]=df_atp[df_atp['Winner']==player].index
    ind_lose[player]=df_atp[df_atp['Loser']==player].index
````

````python
%%time
df_atp['Winner_winperc']=0
df_atp['Loser_winperc']=0
df_atp['Winner_match']=0
df_atp['Loser_match']=0
for player in df_atp['Winner'].unique():
    for index,i in zip(ind_win[player],range(len(ind_win))):
        df_atp.loc[index,'Winner_winperc']=perc_win[player][i]
        df_atp.loc[index,'Winner_match']=matches[player][i]
                                      
for player in df_atp['Loser'].unique():
    for index,i in zip(ind_lose[player],range(len(ind_lose))):
        df_atp.loc[index,'Loser_winperc']=1-perc_lose[player][i]
        df_atp.loc[index,'Loser_match']=matches[player][i]
````

````python
player=[]
x=[]
y=[]
for i in range(3):
    index=randint(0,len(df_atp['Winner'].unique()))
    player1 = df_atp['Winner'].unique()[index]
    player.append(player1)
    y1=df_atp[df_atp['Winner']==player1]['Winner_winperc']
    y.append(y1)
    x1=range(len(y1))
    x.append(x1)

fig,ax = plt.subplots(figsize=(25, 5))
ax.plot(x[0], y[0], label=player[0])
ax.plot(x[1], y[1], label=player[1])
ax.plot(x[2], y[2], label=player[2])
ax.set_xlabel('Matches')  # Add an x-label to the axes.
ax.set_ylabel('Win percentage')  # Add a y-label to the axes.
ax.set_title("Win percentage as a function of played matches")  # Add a title to the axes.
ax.legend();  # Add a legend.
````

````python
#adding the win and lose % for each player in the players DataFrame

players['Win_perc']=a
players['Win_perc'] = players.Win_perc.transform(lambda x: x.fillna(0))
players['Lose_perc']=100-players['Win_perc']
````

````python
#building of the list with the distribution of the wins player in the age segments

young=0
middle=0
old=0
age=df_wta['Age_winner']
for i in age:
    if i<24:
        young+=1
    elif i<31:
        middle+=1
    else:
        old+=1
age=[young,middle,old]
x=['[16-23]','[24-30]','[30+]']
````

````python
#plotting the the distribution of the wins player in the age segments

import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
plt.bar(x, age, width=0.1)
plt.title('\nTrend of victories as a function of the age ranges\n')
plt.xlabel('\nAge range')
plt.ylabel('\nVicotries\n')
plt.show()
````
