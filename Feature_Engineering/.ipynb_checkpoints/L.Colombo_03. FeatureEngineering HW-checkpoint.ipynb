### The dataset

The [Concrete](https://www.kaggle.com/sinamhd9/concrete-comprehensive-strength) dataset contains a variety of concrete formulations and the resulting product's compressive strength, which is a measure of how much load that kind of concrete can bear. For your information, the task for this dataset is to predict a concrete's compressive strength given its formulation.

```python
# Import libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%config InlineBackend.figure_format = "retina"

# set font and plot size to be larger
plt.rcParams.update({'font.size': 20, 'figure.figsize': (20, 13)})
```

```python
url_data = "https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/concrete.csv"

df = pd.read_csv(url_data)
df.head()
```

# KNN - STANDARDIZZAZIONE

## Isolando campione di test per verificare accuratezza modello

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor #regressor perchÃ¨ non ho classi ma un valore, la forza, da calcolare
#r2 score
from sklearn.metrics import r2_score
```

```python
Xs = df.drop(["strength"], axis = 1) #features
ys = df[["strength"]] #label

#ho giÃ  definito prima X e Y

X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(Xs, ys, test_size=0.1, random_state=42) #voglio 10% df di test


#st_scaler = StandardScaler().fit(X)

X_train_prep_st = st_scaler.fit_transform(X_train_s) #df di training con fit e transform
X_test_prep_st = st_scaler.transform(X_test_s) # df di test solo transform
```

```python
#definisco algoritmo KNN

knn_std = KNeighborsRegressor(n_neighbors=5)

knn_std.fit(X_train_prep_st, y_train_s.values.reshape(-1,))  #reshape (-1) per trasformare vettore riga in colonna e viceversa
```

```python
#controllo l'accuratezza delle previsioni

y_pred_s = knn_std.predict(X_test_prep_st)
r2_score(y_test_s, y_pred_s)
```

```python
knn_std.score(X_test_prep_st,y_test_s)
```
