## Exercises

### Exercise - Regression

Consider the [üç´ diabetes dataset](https://scikit-learn.org/stable/datasets/toy_dataset.html#diabetes-dataset) it contains data with $10$ features and a target value representing a quantitative measure of disease progression one year after baseline. Find the _best_ polynomial fit based on the test score for this dataset.

Perform exploratory data analysis, define a suitable performance metric and justify _all_ your choices.
Plot histograms of features distributions and correlation diagrams.

```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_diabetes

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV

import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams.update({'font.size': 20, 'figure.figsize': (20, 13)})
```

```python
dt = load_diabetes()
dt
```

```python
data = pd.DataFrame(dt['data'], columns = dt['feature_names'])
data['target'] = dt['target']
data.info()
```

```python
X=data[data.columns[0:-1]]
y=data[data.columns[-1]]

#split train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

df_train=pd.DataFrame(X_train, columns = data.columns[0:-1])
df_train['target']=y_train

df_test=pd.DataFrame(X_test, columns = data.columns[0:-1])
df_test['target']=y_test
```

```python
g = sns.PairGrid(df_train, height=4)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True);
```

```python
sns.heatmap(df_train.corr(), annot = True, cmap = 'coolwarm')
```

Dai plot si nota come si ha una buona correlazione con le varie feature, quasi tutte sopra 0.3 tranne per 'age', 'sex', 's1' e 's2'.
Questo mi fa capire come un polinomio di grado uno potrebbe essere sufficiente per l'algoritmo di regressione lineare.

```python
fig,axes = plt.subplots(4, 3, constrained_layout = True,figsize = (20,10))
for i,ax in enumerate(axes.ravel()):
    if i > 10:
        fig.delaxes(ax)
        continue
    rgb = np.random.rand(3,)
    sns.histplot( x = df_train.iloc[:,i], kde = True, color = rgb, ax = ax)
```

Visualizzo la distribuzione dei dati per avere un'idea sulla loro distribuzione

```python
corr = df_train.corr().target
corr
```

```python
fig, axes = plt.subplots(4,3, constrained_layout = True, figsize=(20,20))
for i, ax in enumerate(axes.ravel()):
    if i >9:
        fig.delaxes(ax)
        continue
    rgb = np.random.rand(3,)
    sns.regplot(x = df_train[corr.index[i]], y = df_train[corr.index[10]], line_kws={'linewidth':5},color = rgb, ax = ax)
```

Visualizzo le correlazioni e plotto i grafici

```python
k=[i for i in range(1,5)]
scores=[]
degree=[]
rel_err=[]
mean_sqrt=[]
for i in k:
    poly = PolynomialFeatures(degree = i, include_bias = False)
    X_train_ = poly.fit_transform(X_train)
    X_test_ = poly.fit_transform(X_test)
    model = LinearRegression()
    model.fit(X_train_, y_train)
    y_pred=model.predict(X_test_)
    rel=np.mean(abs((y_pred-y_test)/y_test))
    mean=np.sqrt(mean_squared_error(y_pred, y_test))
    scores.append(model.score(X_test_, y_test))
    rel_err.append(rel)
    mean_sqrt.append(mean)
    degree.append(i)
    
print(f"\nConsiderando un polinomio di grado {degree[np.argmax(scores)]}\nsi ottiene un accuratezza pari a {np.round(max(scores),3)*100}%" )

print(f"\nPer ottenere l'errore minimo, pari a {np.round(min(rel_err),2)*100}%,\n(calcolato come media degli errori relativi)\nbisogna utilizzare un polinomio di grado {degree[np.argmin(rel_err)]}" )

print(f"\nPer ottenere l'errore minimo, pari a {np.round(min(mean_sqrt),2)},\n(calcolato usando l'errore quadratico medio)\nbisogna utilizzare un polinomio di grado {degree[np.argmin(mean_sqrt)]}" )
```

Dai grafici gi√† si notava una correlazione lineare, ma ho testato per essere certo che un polinomio di primo grado fosse la scelta giusta per l'algoritmo di regressione lineare.

```python
poly = PolynomialFeatures(degree = 1, include_bias = False)
X_train_ = poly.fit_transform(X_train)
X_test_ = poly.fit_transform(X_test)
model = LinearRegression()
model.fit(X_train_, y_train)
y_pred=model.predict(X_test_)
err_rel_lr=abs((y_pred-y_test)/y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
k=[i for i in range(2,30) if i%2!=0]
dist=['euclidean','cosine','manhattan']
```

```python
best_scores=[]
best_rel_err=[]
best_mean_sqrt=[]
best_k=[]
for j in dist:
    scores=[]
    rel_err=[]
    mean_sqrt=[]
    for i in k:
        knn = KNeighborsRegressor(n_neighbors=i, metric=j)
        knn.fit(X_train,y_train)
        scores.append(knn.score(X_test, y_test))
        y_pred=knn.predict(X_test)
        rel=np.mean(abs((y_pred-y_test)/y_test))
        mean=np.sqrt(mean_squared_error(y_pred, y_test))
        rel_err.append(rel)
        mean_sqrt.append(mean)
    
    best_scores.append(np.max(scores))
    best_rel_err.append(np.min(rel_err))
    best_mean_sqrt.append(np.min(mean_sqrt))
    best_k.append(k[np.argmax(scores)])
    plt.plot(k,scores,label=j)
    
plt.xlabel('N¬∞ of neighbors')  # Add an x-label to the axes.
plt.ylabel('Accuracy')  # Add a y-label to the axes.
plt.title("Accuracy trend")
plt.legend()

print(f"La scelta migliore risulta utilizzando la distanza {dist[np.argmax(best_scores)]},\ncon un k={best_k[np.argmax(best_scores)]} e un'accuratezza pari a {np.round(np.max(best_scores)*100,2)}%")
print(f"\nL'errore relativo √® pari a {np.round(min(best_rel_err),2)*100}%\ne l'errore quadradico medio √® pari a {np.round(min(best_mean_sqrt),2)}")
```

Per confronto ho utilizzato anche l'algoritmo knn, per verificare se l'accuratezza del modello cambia.
Effettivamente sembra che il modello knn sia pi√π accurato, di circa il 6% in pi√π, rispetto al modello di regressione lineare.
L'errore quadradico medio √® inferiore per il knn rispetto al modello di regressione lineare.
Mentre l'errore relativo medio non sembra variare molto.

```python
knn = KNeighborsRegressor(n_neighbors=13, metric='cosine')
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
err_rel_knn=abs((y_pred-y_test)/y_test)
```

```python
parameters = {'n_neighbors':[i for i in range(1,30) if i%2!=0],
              'weights':['uniform', 'distance'],
              'metric':['euclidean','manhattan', 'cosine']}
knn = KNeighborsRegressor()
grid = GridSearchCV(knn, parameters) 
grid.fit(X_train, y_train)
print('KNN score =', grid.best_estimator_.score(X_test, y_test))
print('KNN best parameters =', grid.best_params_)

hmknn = pd.pivot_table(pd.DataFrame(grid.cv_results_),
									values='mean_test_score',
									index='param_n_neighbors',
									columns='param_metric')

sns.heatmap(hmknn, cmap = 'magma', annot = True)

#Calcolo gli scarti dei dati dal modello
err_knn = (grid.predict(X_test) - y_test)/y_test
```

Ho utilizzato la gridsearch per verificare quali siano i parametri migliori da utilizzare per il modello knn ,il risultato √® che con il codice precedente ho ottenuto del parametri differenti e sopratutto un'accuratezza maggiore rispetto alla ricerca dei parametri tramite la gridsearch.

Per provare a migliorare la previsione ho provato a modificare i dati. Infatti ho moltiplicato ciascuna colonna per la corrispondete correlazione con la colonna target, successivamente ho sommato tutti i prodotti e ho ottenuto una nuova colonna.
Quindi ho utilizzato di nuovo i modelli utilizzati in precedenza per verificare se questo porta a un miglioramento o meno.

```python
corr=data.corr().target[:-1]
data_corr=X*corr
data.drop('target', axis = 1, inplace = True)
data['sum']=data_corr.age+data_corr.sex+data_corr.bmi+data_corr.bp+data_corr.s1+data_corr.s2+data_corr.s3+data_corr.s4+data_corr.s5+data_corr.s6
data['target']=y
```

```python
X=data[data.columns[0:-1]]
y=data[data.columns[-1]]

#split train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

df_train=pd.DataFrame(X_train, columns = data.columns[0:-1])
df_train['target']=y_train

df_test=pd.DataFrame(X_test, columns = data.columns[0:-1])
df_test['target']=y_test
```

```python
k=[i for i in range(1,5)]
scores=[]
degree=[]
rel_err=[]
mean_sqrt=[]
for i in k:
    poly = PolynomialFeatures(degree = i, include_bias = False)
    X_train_ = poly.fit_transform(X_train)
    X_test_ = poly.fit_transform(X_test)
    model = LinearRegression()
    model.fit(X_train_, y_train)
    y_pred=model.predict(X_test_)
    rel=np.mean(abs((y_pred-y_test)/y_test))
    mean=np.sqrt(mean_squared_error(y_pred, y_test))
    scores.append(model.score(X_test_, y_test))
    rel_err.append(rel)
    mean_sqrt.append(mean)
    degree.append(i)
    
print(f"\nConsiderando un polinomio di grado {degree[np.argmax(scores)]}\nsi ottiene un accuratezza pari a {np.round(max(scores),3)*100}%" )

print(f"\nPer ottenere l'errore minimo, pari a {np.round(min(rel_err),2)*100}%,\n(calcolato come media degli errori relativi)\nbisogna utilizzare un polinomio di grado {degree[np.argmin(rel_err)]}" )

print(f"\nPer ottenere l'errore minimo, pari a {np.round(min(mean_sqrt),2)},\n(calcolato usando l'errore quadratico medio)\nbisogna utilizzare un polinomio di grado {degree[np.argmin(mean_sqrt)]}" )
```

Con il modello di regressione lineare non sembra ci siano miglioramenti, n√® peggioramenti, anzi i risultati sembrano invariati!

```python
poly = PolynomialFeatures(degree = 1, include_bias = False)
X_train_ = poly.fit_transform(X_train)
X_test_ = poly.fit_transform(X_test)
model = LinearRegression()
model.fit(X_train_, y_train)
y_pred=model.predict(X_test_)
err_rel_lr_corr=abs((y_pred-y_test)/y_test)
```

```python
k=[i for i in range(2,30) if i%2!=0]
dist=['euclidean','cosine','manhattan']
```

```python
best_scores=[]
best_rel_err=[]
best_mean_sqrt=[]
best_k=[]
for j in dist:
    scores=[]
    rel_err=[]
    mean_sqrt=[]
    for i in k:
        knn = KNeighborsRegressor(n_neighbors=i, metric=j)
        knn.fit(X_train,y_train)
        scores.append(knn.score(X_test, y_test))
        y_pred=knn.predict(X_test)
        rel=np.mean(abs((y_pred-y_test)/y_test))
        mean=np.sqrt(mean_squared_error(y_pred, y_test))
        rel_err.append(rel)
        mean_sqrt.append(mean)
    
    best_scores.append(np.max(scores))
    best_rel_err.append(np.min(rel_err))
    best_mean_sqrt.append(np.min(mean_sqrt))
    best_k.append(k[np.argmax(scores)])
    plt.plot(k,scores,label=j)
    
plt.xlabel('N¬∞ of neighbors')  # Add an x-label to the axes.
plt.ylabel('Accuracy')  # Add a y-label to the axes.
plt.title("Accuracy trend")
plt.legend()

print(f"\nLa scelta migliore risulta utilizzando la distanza {dist[np.argmax(best_scores)]},\ncon un k={best_k[np.argmax(best_scores)]} e un'accuratezza pari a {np.round(np.max(best_scores)*100,2)}%")
print(f"\nL'errore relativo √® pari a {np.round(min(best_rel_err),2)*100}%\ne l'errore quadradico medio √® pari a {np.round(min(best_mean_sqrt),2)}")
```

Il modello knn sembra mostrare un leggero miglioramento rispetto al precedente, infatti l'accuratezza √® aumentato anche se di poco, mentre gli errori risultano essere inferiori.
